```julia
using Gadfly
using DiffEQ
```

```julia
function f(t, y)
    ydot = similar(y)
    ydot[1] = y[2]*y[3]
    ydot[2] = -y[1]*y[3]
    ydot[3] = -0.51*y[1]*y[2]
    ydot
end
```

```julia
t = collect(linspace(0.0, 12.0, 100));
sol = ode45(f, [0.0, 1.0, 1.0], t; reltol = 1e-4, abstol = 1e-4);
```

```julia
plot(
layer(x = sol.x, y = sol.y[:, 1], Geom.line),
layer(x = sol.x, y = sol.y[:, 2], Geom.line, Theme(default_color = colorant"red")),
layer(x = sol.x, y = sol.y[:, 3], Geom.line, Theme(default_color = colorant"yellow"))
)
```
